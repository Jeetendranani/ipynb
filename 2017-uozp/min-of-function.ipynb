```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib notebook
```

Find a minimum of a single parameter function.

```python
def J(a):
    return (a-5)**2+3
```

```python
a = np.arange(0,10,0.1)
y = J(a)
```

```python
plt.plot(a, y, "k-");
```

```python
def derivative(f, a, eps=1e-3):
    return (f(a+eps) - f(a-eps)) / (2*eps)
```

```python
ai = 8
for _ in range(10):
    ai = ai - 0.2 * derivative(J, ai)
    print("%.2f" % ai)
```

Find a minimum of a function with two parameters.

```python
from mpl_toolkits.mplot3d import axes3d
```

```python
def J3(a, b):
    return (a - 1)**2 + (b + 1)**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([J3(x, y) for x, y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_wireframe(X, Y, Z, rstride=20, cstride=20);
```

```python
def grad(f, point, e=1e-3):
    return np.array([(f(*(point+eps)) - f(*(point-eps)))/(2*e)
                     for eps in np.identity(len(point)) * e])
```

```python
grad(J3, np.array([0, 0]))
```

```python
thetas = np.array([0, 0])
for _ in range(10):
    thetas = thetas - 0.2 * grad(J3, thetas)
    print(thetas)
```

We can extend the concept with finding a minimum of a function to function with arbitrary number of parameters. Following is a function with with three parameters.

```python
def J33(theta0, theta1, theta2):
    return (theta0 - 1)**2 + (theta1 + 1)**2 + (theta2-42)**2
```

```python
thetas = np.array([0, 0, 0])
for _ in range(10):
    thetas = thetas - 0.3 * grad(J33, thetas)
    print(thetas)
```
